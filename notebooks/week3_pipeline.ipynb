```python
# week3_pipeline.ipynb
# ===================================================
# WEEK-3 PIPELINE: Query → Retrieval → Insight → Summary
# ===================================================

# 1️⃣ Imports
import sys, os
import pandas as pd
import numpy as np
import json

# Add src to path
sys.path.insert(0, os.path.abspath(".."))
from src.retrieval import retrieve_reviews, get_cluster_distribution
from src.insight_generation import generate_insight, summarize_reviews

# ===================================================
# 2️⃣ Load Data
# ===================================================
df = pd.read_csv("../analysis/reviews_with_clusters.csv")
cluster_keywords_df = pd.read_csv("../analysis/cluster_keywords.csv")

print(f"Data loaded: {len(df)} reviews")

# ===================================================
# 3️⃣ Example Query
# ===================================================
query = "bad coffee taste"

# Step 1: Retrieve top reviews (pass raw query string, not embedding)
results = retrieve_reviews(query, k=10)  # list of dicts

# Step 2: Extract indices
indices = [item["index"] for item in results]

# Step 3: Get cluster distribution
cluster_info = get_cluster_distribution(indices, df)

# Step 4: Extract review texts
retrieved_reviews = [item["review_text"] for item in results]

# ===================================================
# 4️⃣ Generate Insight
# ===================================================
insight = generate_insight(retrieved_reviews, cluster_info, cluster_keywords_df)

# ===================================================
# 5️⃣ Summarize Reviews
# ===================================================
summary = summarize_reviews(retrieved_reviews)

# ===================================================
# 6️⃣ Output Results
# ===================================================
print("=== QUERY ===")
print(query)
print("\n=== INSIGHT ===")
print(json.dumps(insight, indent=2))
print("\n=== SUMMARY ===")
print(json.dumps(summary, indent=2))
```
